```python
import pandas as pd
data = pd.read_csv('HomeC.csv')
data.head()
```

```python
from sdv.tabular import GaussianCopula
model = GaussianCopula()
model.fit(data)
```

```python
synthetic_data = model.sample(5000000)
synthetic_data.head()
```
