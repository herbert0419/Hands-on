```python
meter.columns = [col.replace(' [kW]', '') for col in meter.columns]
meter.columns
```

```python
meter['sum_Furnace'] = meter[['Furnace 1','Furnace 2']].sum(axis=1)
meter['avg_Kitchen'] = meter[['Kitchen 12','Kitchen 14','Kitchen 38']].mean(axis=1)
```

```python
meter = meter.drop(['Kitchen 12','Kitchen 14','Kitchen 38'], axis=1)
meter = meter.drop(['Furnace 1','Furnace 2'], axis=1)
meter.columns
```

```python
meter['Date & Time'].head()
```

```python
time_index = pd.date_range('2016-01-01 05:00', periods=len(meter),  freq='min')  
time_index = pd.DatetimeIndex(time_index)
meter = meter.set_index(time_index)
meter = meter.drop(['Date & Time'], axis=1)
meter.iloc[np.r_[0:5,-5:0]].iloc[:,0]
```
