```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()

# Define the input variables x and y
x = tf.placeholder(tf.float32)
y = tf.placeholder(tf.float32)

# Define the equation
z = x**3 + y**3 + 20*x**2*y + 30*y**2*x + 10

# Create a session to evaluate the equation
with tf.Session() as sess:
    result1 = sess.run(z, feed_dict={x: 10, y: 20})
    result2 = sess.run(z, feed_dict={x: 23, y: 34})
    result3 = sess.run(z, feed_dict={x: 252, y: 423})
    print(result1) 
    print(result2)
    print(result3)  
```
